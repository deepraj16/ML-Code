```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt  
import seaborn as sns  
```

```python
df=pd.read_csv("data.csv")
```

```python
df.head(3)
```

```python
X=df.iloc[:,:-1]
y=df['target'].values
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=42,random_state=23)
```

```python
from sklearn.linear_model import LinearRegression 
lr= LinearRegression() 
lr.fit(X_train,y_train)
y=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score 

residual=y_test-y
r2_score(y_test,y)
```
